> ### General Analysis

 *  ### <u> Density of Life Expectancy on the basis on status </u>

```python
df_developed = df.loc[df['Status'] == 'Developed'].copy()
df_developing = df.loc[df['Status'] == 'Developing'].copy()
```

```python
sns.kdeplot(df_developed['Life expectancy'] , 
            label = 'Developed',
            fill = True,
            color = colors[0],
            alpha = 0.8)

sns.kdeplot(df_developing['Life expectancy'] , 
            label = 'Developing',
            fill = True,
            color = colors[1],
            alpha = 0.8)

plt.legend(loc = 'upper left')
plt.ylabel('Density')
plt.xlim(30,90)
plt.xlabel('Life Expectancy(years))')
plt.title('Life Expectancy Based On Status \n' , fontsize=18);
```

#### Inference : 'Developed' column has more value towards right depicting life expectancy for developed countries are more than developing countries

* ### <u> Life Expectancy over the Years </u>

```python
fig,ax = plt.subplots(figsize = (14,8))

ax.plot(df_developed.groupby('Year')['Life expectancy'].mean(),
        label = 'Developed',
        color = colors[0],
        linewidth = 4)

ax.fill_between(df_developed.groupby('Year')['Life expectancy'].mean().index,
                df_developed.groupby('Year')['Life expectancy'].mean().values,
                color = colors[0],
                alpha=0.8)

ax.plot(df_developing.groupby('Year')['Life expectancy'].mean(),
        label = 'Developing',
        color = colors[1],
        linewidth = 4)

ax.fill_between(df_developing.groupby('Year')['Life expectancy'].mean().index,
                df_developing.groupby('Year')['Life expectancy'].mean().values,
                color = colors[1],
                alpha=0.8)


plt.legend(loc='upper left', fontsize=12)
ax.set_xlabel('Year')
ax.set_ylabel('Average Life Expectancy (years)')

# Set y-axis limits
ax.set_ylim(60,82)

ax.set_title('Average Life Expectancy over the years in developed and developing countries\n' , fontsize = 18);
```

#### Inference : Growth in Life Expectancy for both developing and developed is positive over the year and gradually increasing

* ### <u> Correlation between features </u>

```python
#Plot Matrix of Correlation
plt.figure(figsize=(17,17))
plt.rcParams.update({'font.size': 14})

heatmap_colors = [colors[1], '#d6d5c9', '#FE0000']

corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(df.corr(), 
            mask=mask,
            center=0,
            annot=True,
            fmt='.2f',
            cmap=heatmap_colors,
            square=True,
            linewidths=.2,
            cbar_kws={"shrink": .6})

plt.title('Features Correlation Matrix Heatmap', fontsize=25);
```

- ####  The correlation shows that target have high correlation with Schooling , Income compostion of resources, BMI and high negative correlation with HIV/AIDS , Adult Mortality.
- ####  Schooling and Income compostition of resources also have high correlation between them , also HIV/AIDS.
- ####  Other factors are also highly correlated like Percentage expenditure & GDP , Diphtheria & Polio , Income compostion of resources & BMI , Measles & Infant death.

It may be due to countries having good financial condition , population might consuming more alcohol but it does not justify the fact of life expectancy ( you may be owning a supercar but owning a supercar does not mean you will have greater life expectancy).
